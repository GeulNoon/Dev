```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
morph_list=[]
for i in txt:
 morphs = kkma.morphs(i)

 for morph in morphs:
   morph_list.append(morph)
```

```python
morph_list
```

```python
from collections import Counter
count = Counter(morph_list)
```

```python
count
```

```python
count.most_common()
```

```python
for i in morph_list:
  f=open('kobartkkmamorph.txt','a',encoding='cp949')
  f.write(i+'\n')
  f.close()
```
